# Libraries

```python
# importing libraries
import pandas as pd
import requests
from urllib.request import urlretrieve as retrevie
```

# TGE.pl

```python
# downloading data and saving as csv file
url = 'https://tge.pl/energia-elektryczna-rdn'
df0 = pd.read_html(url)[2]
df0.columns = [f'{x[0]}_{x[1]}' if x[1] else x[0] for x in df0.columns] # Concatenate the headers from two levels of the MultiIndex
df0.rename(columns={"Unnamed: 0_level_0_Czas": "Czas"}, inplace=True)
df0.to_csv("TGE_data.csv")
df0.head(5)
```

# PSE.pl

```python
# defining variables
data_od = "20230205"
data_do = "20230206"
url1 = f"https://www.pse.pl/getcsv/-/export/csv/PL_GEN_MOC_JW_EPS/data_od/{data_od}/data_do/{data_do}"
url2 = f"https://www.pse.pl/getcsv/-/export/csv/PL_WYK_KSE/data_od/{data_od}/data_do/{data_do}"
filename1 = f"PL_GEN_MOC_JW_od_{data_od}_do_{data_do}.csv"
filename2 = f"PL_WYK_KSE_od_{data_od}_do_{data_do}.csv"
urls_to_download = [url1, url2]
files_to_download = [filename1, filename2]
```

```python
# saving csv files with data
for url, file in zip(urls_to_download, files_to_download):
    retrevie(url=url, filename=file)
```

```python
# storing data into two dataframes
df1 = pd.read_csv(f"./{filename1}", sep=";", decimal=",", encoding="cp1250")
df2 = pd.read_csv(f"./{filename2}", sep=";", decimal=",", encoding="cp1250")
```
